```python
import trollius as asyncio
from trollius import From

from psychopy import visual
import pygame, pyglet
win=visual.Window(size=(800,600), fullscr=False, winType='pyglet', allowStencil=True, waitBlanking=True)


@asyncio.coroutine
def flipwin_co(win):
    win.flip()
    print('flipped!)')

# flipwin2(win)

@asyncio.coroutine
def flipwin(win):
    for x in range(0,100):
        print('before')
        yield From(flipwin_co(win))
        print('after')
        

@asyncio.coroutine
def fast_stuff():
    for x in range(0,2000):
        print('1')
        yield From(asyncio.sleep(0))


@asyncio.coroutine
def factorial(name, number):
    f = 1
    for i in range(2, number + 1):
        print("Task %s: Compute factorial(%d)..." % (name, i))
        yield From(asyncio.sleep(1))
        f *= i
    print("Task %s completed! factorial(%d) is %d" % (name, number, f))

loop = asyncio.get_event_loop()
tasks = [
    #asyncio.async(factorial("A", 8)),
    #asyncio.async(factorial("B", 3)),
    #asyncio.async(factorial("C", 4)),
    asyncio.async(flipwin(win)),
    asyncio.async(fast_stuff()),
]
loop.run_until_complete(asyncio.wait(tasks))
loop.close()
```
